```python
cormatrix=df.corr()
cormatrix
```

```python
cormatrix *=np.tri(*cormatrix.values.shape,k=-1).T# tri函数生成一个矩阵，在某对角线以下元素全为1，其余全为0，对角线由参数k控制，它是一个整数值，默认情况下为0。如果k> 0的值，则表示对角线位于主对角线上方，反之亦然
cormatrix=cormatrix.stack()
cormatrix
```

```python
cormatrix=cormatrix.reindex(cormatrix.abs().sort_values(ascending=False).index).reset_index()
cormatrix[:10]
```

```python
from sklearn.preprocessing import StandardScaler#标准差标准化:转化函数：x = (x-mean) / std
target=df['price']
regressor=[x for x in df.columns if x not in ['price']]
features=df.loc[:,regressor]
#features
stand_scaler=StandardScaler()
num=['symboling', 'normalized-losses', 'volumn', 'horsepower', 'wheel-base',
       'bore', 'stroke','compression-ratio', 'peak-rpm']
features[num]=stand_scaler.fit_transform(features[num])
features
```

```python
#对分类属性进行独热编码
classes = ['make', 'fuel-type', 'aspiration', 'num-of-doors', 
           'body-style', 'drive-wheels', 'engine-location',
           'engine-type', 'num-of-cylinders', 'fuel-system']

dummies=pd.get_dummies(features[classes])#pandas中的独热编码方法
dummies
features=features.join(dummies).drop(classes,axis=1)
features.head()
```

```python
#Lasso回归，不同的alpha得到不同的结果
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import Lasso,LassoCV,LogisticRegression
from sklearn.ensemble import RandomForestRegressor
seed = 123
xtrain,xtest,ytrain,ytest=train_test_split(features,target,test_size=0.3,random_state=seed)
print('train_shape',xtrain.shape,'test_shape',xtest.shape)
```

```python
alphas = 2. ** np.arange(2, 12)
scores = np.empty_like(alphas)
scores
```

```python
for i, a in enumerate(alphas):
    lasso = Lasso(random_state = seed)
    lasso.set_params(alpha = a)
    lasso.fit(xtrain, ytrain)
    scores[i] = lasso.score(xtest, ytest)
scores
```

```python
lassocv=LassoCV(cv=10,random_state=seed)
lassocv.fit(features,target)
lassocv_score=lassocv.score(features,target)
lassocv_alpha=lassocv.alpha_
print(lassocv_score)
print(lassocv_alpha)
```

```python
model_1 = LassoCV(alphas = alphas, cv = 10, random_state = seed)
#model_1=Lasso(alpha=lassocv_alpha,random_state=seed)
model_1.fit(xtrain,ytrain)
ypred=model_1.predict(xtest)
model_1.score(xtest,ytest)
#t=pd.DataFrame(model_1.coef_)
#t1=pd.DataFrame
for i in range(0,66):#66是列数
    print(model_1.coef_[i],xtrain.columns[i])
#print(xtrain.columns)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
def MSE(y_true,y_pred):
    mse = mean_squared_error(y_true, y_pred)
    print('MSE: %2.3f' % mse)
    return mse

def R2(y_true,y_pred):    
    r2 = r2_score(y_true, y_pred)
    print('R2: %2.3f' % r2)     
    return r2

MSE(ytest, ypred); R2(ytest, ypred);
```
