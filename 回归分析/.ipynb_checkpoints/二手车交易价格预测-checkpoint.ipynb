    （1）保持方法：将原始数据分成两个不相交的集合，训练集和测试集，train_test_split()方法（不放回抽样）
    （2）K折交叉验证：将数据分成大小相同的K份，每次运行时，选取其中一份做检验集，其他都做训练集，重复K次，总误差=K次误差之和，
    KFOLD划分数据集，cross_val_score()交叉验证（不放回抽样）
    （3）自助法：有放回抽样，.632自助法

    （4）boosting：

    （6）stacking:

5.不平衡类问题（《数据挖掘导论》P180）

     比如产品检测中不合格产品的数量远远低于合格产品的数量等这种稀有类的正确分类比多数类的正确分类更有价值的分类问题。
     （1）除准确度之外的其他度量
     真正：TP；假负：FN；假正：FP；真负：TN
     精度：p=TP/(TP+FP):确定在分类器断言为正类的那部分记录中实际为正类的记录所占的比例
     召回率：r=TP/(TP+FN)：度量被分类器正确预测的正样本的比例
     F1度量：精度和召回率的调和均值
     （2）
     ROC曲线
     AUC：ROC曲线下方的面积，如果模型是完美的，则AUC=1，如果仅是简单的随机猜测，则AUC=0.5，我们的分类模型的AUC大于0.5就是有效的。如果一个模型好于另一个，则它的AUC较大
     （3）代价敏感学习方法
     （4）基于抽样的方法

```python
#一些从常识来看和价格无关的特性可以剔除掉，由上面的相关性矩阵也能看出
num_col = df.select_dtypes(exclude='object').columns
num_col
feature_col = [col for col in num_col if col not in ['SaleID','name','regDate','creatDate','seller','offerType','price']]
#feature_col = [col for col in feature_col if 'Type' not in col]
df[feature_col].head()
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.linear_model import LinearRegression 
from sklearn.linear_model import Ridge,Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from lightgbm.sklearn import LGBMRegressor
from xgboost.sklearn import XGBRegressor
from sklearn.metrics import mean_absolute_error#MAE越小，模型越准确
xdata=df[feature_col]
ydata=df['price']


fig,ax=plt.subplots(1,2,figsize=(12,8))
ydata_ln = np.log(ydata+1)#价格字段不符合正态分布，基本都是集中在0-10000之间,进行对数变换之后基本是正态分布了
ax[0].hist(ydata)
ax[1].hist(ydata_ln)
print(ydata_ln[:5])

xtrain,xtest,ytrain,ytest=train_test_split(xdata,ydata_ln,test_size=0.3)
print('xtrain:',xtrain.shape)
print('ytrain:',ytrain.shape)
print('xtest:',xtest.shape)
print('ytest:',ytest.shape)
```

```python
linear = LinearRegression()
linear.fit(xtrain,ytrain)
ypred_linear = linear.predict(xtest)
linear_mae = mean_absolute_error(ytest,ypred_linear)
print(linear_mae)
#ridge = ridge_regression()
fig,ax = plt.subplots(1,2,figsize=(12,8))
ax[0].scatter(ytest,ypred_linear)
ax[1].hist(ypred_linear)
```

```python
ridge = Ridge()
ridge.fit(xtrain,ytrain)
ypred_ridge = ridge.predict(xtest)
mse_ridge = mean_absolute_error(ytest,ypred_ridge)
print(mse_ridge)
plt.scatter(ytest,ypred_ridge)
```

```python
dtr = DecisionTreeRegressor()
dtr.fit(xtrain,ytrain)
ypred_drt = dtr.predict(xtest)
mse_dtr = mean_absolute_error(ytest,ypred_drt)
print(mse_dtr)
fig,ax = plt.subplots(1,2,figsize=(12,8))
ax[0].scatter(ytest,ypred_drt)
ax[1].hist(ypred_drt)
```

```python
rf = RandomForestRegressor(n_estimators=10)
rf.fit(xtrain,ytrain)
ypred_rf = rf.predict(xtest)
mse_rf = mean_absolute_error(ytest,ypred_rf)
print(mse_rf)
```

```python
#发现随机森林相比岭回归，线性回归要好很多，但是这样一个一个试算法其实很麻烦，有一种方法很精妙，把所有模型放在一个字典中，
#然后分别进行交叉验证比较，直接用箱线图看各个模型的的绝对值误差
from sklearn.model_selection import KFold
num_fold = 10
seed=7
#把所有模型写到一个字典中
models = {}
models['LR']=LinearRegression()
models['Ridge']=Ridge()
models['Lasso']=Lasso(alpha=0.01)
models['DecisionTree']=DecisionTreeRegressor()
models['RandomForest']=RandomForestRegressor(n_estimators=10)
models['GradientBoosting']=GradientBoostingRegressor()
models['XGB']= XGBRegressor(n_estimators = 100, objective='reg:squarederror')
models['LGB']=LGBMRegressor(n_estimators=100)

results=[]
for key in models:
    kfold = KFold(n_splits=num_fold,random_state=seed)
    cv_result = cross_val_score(models[key],xdata,ydata_ln,cv=kfold,scoring='neg_mean_absolute_error')
    results.append(cv_result)
    print('%s:%f(%f)' % (key,cv_result.mean(),cv_result.std()))


fig1 = plt.figure(figsize=(15, 10))
fig1.suptitle('Algorithm Comparison')
ax = fig1.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(models.keys())
plt.show()
```

```python

from sklearn.model_selection import GridSearchCV
cv_params = {'n_estimators': [400, 500, 600, 700, 800]}
params = {'learning_rate':0.1,'seed': 7, 'gamma': 0.1, 
                            'reg_alpha': 1, 
                            'reg_lambda': 1, 
                            'eval_metric':'auc'}
model = XGBRegressor(**params)
optimized_GBM = GridSearchCV(estimator=model, param_grid=params, scoring='neg_mean_absolute_error', cv=5, verbose=1, n_jobs=4)
optimized_GBM.fit(xtrain,ytrain)
ypred_xgb_adjust = optimized_GBM.predict(xtest)
mse_xgb_adjust = mean_absolute_error(ytest,ypred_xgb_adjust)
```
